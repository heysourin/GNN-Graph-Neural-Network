## **Neighbors and Adjacency:**

```python
G = nx.DiGraph()
G.add_edge('a', 'b', weight=1)
G.add_edge('c', 'b', weight=5)
G.add_edge('m', 'n', weight=25)
G.add_edge('m', 'b', weight=50)
nx.draw(G)
print(nx.is_weighted(G))
print(nx.is_directed(G))
print(G.order())
print(G.number_of_edges())
print(G.number_of_nodes())
print(G.edges)
```

```python
G.order()
```

```python
print([n for n in G.neighbors('a')])
```

```python
print([n for n in G.neighbors('c')])
```

# **Multiple Edge Attributes**

```python
import numpy as np

G = nx.Graph()
G.add_edge(0, 1, weight=10)
G.add_edge(1, 2, cost=5)
G.add_edge(2, 3, weight=3, cost=-4.0)
dtype = np.dtype([("weight", int), ("cost", float)])
A = nx.to_numpy_array(G, dtype=dtype, weight=None)
print("weight --------")
print(A["weight"])
print("cost --------")
print(A["cost"])
nx.draw(G)
print(G.edges)
print(G.nodes)
```

```python
G = nx.Graph()
G.add_edges_from(
  [
    (0, 1, {'weight': 1.0}),
    (1, 2, {'weight': 2.0}),
    (2, 0, {'weight': 1.0}),
  ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
d = nx.to_dict_of_dicts(G)
print(d)
print(d[1][2]['weight'])
```

```python
G = nx.MultiGraph()
G.add_edge(0, 1, key='a', weight=1.0)
G.add_edge(0, 1, key='b', weight=5.0)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
d = nx.to_dict_of_dicts(G)
print(d)
print(d[0][1]['b']['weight'])
```

```python
G = nx.Graph(
  [
    ("A", "B", {"cost": 1, "weight": 7}),
    ("C", "E", {"cost": 9, "weight": 10}),
  ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G)
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])
print(df)
print("--------------------")
df[["source", "target", "cost", "weight"]]
```

```python
G = nx.Graph(
    [
        ("A", "B", {"cost": 1, "weight": 7}),
        ("C", "E", {"cost": 9, "weight": 10}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G)
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])
print(df)
print("--------------------")
df[["source", "target", "cost", "weight"]]
print("--------------------")
A = nx.adjacency_matrix(G)
print(A.todense())
```

```python
G = nx.MultiGraph([('A', 'B', {'cost': 1}), ('A', 'B', {'cost': 9})])
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=['A', 'C'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=['B'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
```
